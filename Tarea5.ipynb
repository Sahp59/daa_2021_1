```
import time
# 4
def ejemplo1( n ):
    inicio = time.time() #Inicia Conteo
    c = n + 1
    d = c * n
    e = n * n
    total = c + e - d
    fin = time.time()   #Finaliza Conteo
    tiempo_total = fin - inicio     #Calculamos el tiempo
    print(tiempo_total,"segundos")  #Imprimimos el tiempo

#1+n^2
def ejemplo2( n ):
    inicio = time.time() #Inicia Conteo
    contador = 0
    for i in range( n ) :
        for j in range( n ) :
            contador += 1
    fin = time.time()   #Finaliza Conteo
    tiempo_total = fin - inicio     #Calculamos el tiempo
    print(tiempo_total,"segundos")  #Imprimimos el tiempo
    return contador

#102
def ejemplo3( n ): # n=4
    inicio = time.time() #Inicia Conteo
    x = n * 2 # x = 8
    y = 0   # y = 0
    for m in range( 100 ): #3
        y = x - n   # y = 4
    fin = time.time()   #Finaliza Conteo
    tiempo_total = fin - inicio     #Calculamos el tiempo
    print(tiempo_total,"segundos")  #Imprimimos el tiempo
    return y

#3
def ejemplo4( n ):
    inicio = time.time() #Inicia Conteo
    x = 3 * 3.1416 + n
    y = x + 3 * 3 - n
    z = x + y
    fin = time.time()   #Finaliza Conteo
    tiempo_total = fin - inicio     #Calculamos el tiempo
    print(tiempo_total,"segundos")  #Imprimimos el tiempo
    return z

#1+x
def ejemplo5( x ):
    inicio = time.time() #Inicia Conteo
    n = 10
    for j in range( 0 , x , 1 ):
        n = j + n
    fin = time.time()   #Finaliza Conteo
    tiempo_total = fin - inicio     #Calculamos el tiempo
    print(tiempo_total,"segundos")  #Imprimimos el tiempo
    return n

#2+n^3
def ejemplo6( n ):
    inicio = time.time() #Inicia Conteo
    data=[[[1 for x in range(n)] for x in range(n)]
          for x in range(n)]
    suma = 0
    for d in range(n):
        for r in range(n):
            for c in range(n):
                suma += data[d][r][c]
    fin = time.time()   #Finaliza Conteo
    tiempo_total = fin - inicio     #Calculamos el tiempo
    print(tiempo_total,"segundos")  #Imprimimos el tiempo
    return suma

def main(n):
    print("tiempo de ejecucion ejemplo 1")
    ejemplo1(n)
    print("tiempo de ejecucion ejemplo 2")
    ejemplo2(n)
    print("tiempo de ejecucion ejemplo 3")
    ejemplo3(n)
    print("tiempo de ejecucion ejemplo 4")
    ejemplo4(n)
    print("tiempo de ejecucion ejemplo 5")
    ejemplo5(n)
    print("tiempo de ejecucion ejemplo 6")
    ejemplo6(n)

print("Todas la funciones recorreran valores de 100 a 1100 de 100 en 100")
for entrada in range(100,1100,100):
    print(entrada)
    main(entrada)
```
